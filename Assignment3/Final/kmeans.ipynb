```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import statistics as stats
from kneed import KneeLocator

# Function to generate data points randomly
def generate_data(n_samples, n_features, random_state, centers):
    data, labels = make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, random_state=random_state)
    return data, labels

# Function to depict the scatter plot along with centroids
def plot(x, centroids):
    centroids = np.array(centroids)
    plt.scatter(x[:, 0], x[:, 1])
    plt.scatter(centroids[:, 0], centroids[:, 1])
    plt.show()

# Function to calculate new centroids 
def find_new_Uk(data):
    new_Uk = []
    # Iterate through all the labels from data
    for label in data:
        if len(label) > 0:
            # Calculate the mean of all points and generate new coordinates for new Uk/centroid
            x_mean = stats.mean([elem[0] for elem in label])
            y_mean = stats.mean([elem[1] for elem in label])
            new_Uk.append([x_mean, y_mean])
    return new_Uk

# Function to perform clustering using KMeans algorithm
def k_means(x, K, n_init):
    # x: input data
    # K: number of centroids/clusters
    # n_init: the number of initial guesses for the centroids
        
    #Creating list of sets for storing the centroids and the data points assigned to each of those.
    labels = [set() for i in range(1,K+1)]

    x_size=len(x)
    # rand number for init centroids,
    i = 0

    new_centriod_vals = []
    centroid_values = []
    for n in range(n_init):
        i += 1
        print(f'step {i}')
        
        # Generate initial points for centroids randomly
        if not new_centriod_vals:
            rand_num_from_x = [random.randint(0, x_size - 1) for i in range(K)] # K random numbers
            centroid_values = [x[rand_num_from_x[i]] for i in range(len(rand_num_from_x))] # Initialise centroid for K clusters
            
        else:
            # Compare the newly generated centroids with the previous value and
            # if they are same(or they converge) then stop the iteration and return
            if np.array_equal(new_centriod_vals,centroid_values):
                print('Ending')
                return new_centriod_vals, labels
            else:
            # if they do not match then assign new centroid values to current centroid    
                centroid_values = new_centriod_vals

        # get euclidean distance of N points from u_k, assign xi to the cloest u_k
        # get x_i
        for x_i in x_data:
            eucid_d_x_i_centroids = []
            for cen_i in centroid_values:
                eucid_d = math.sqrt((cen_i[0] - x_i[0]) ** 2 + (cen_i[1] - x_i[1]) ** 2)
                eucid_d_x_i_centroids.append(eucid_d)
            closet_c_to_xi = np.argmin(eucid_d_x_i_centroids)
            labels[int(closet_c_to_xi)].add((x_i[0], x_i[1]))
            
        new_centriod_vals = find_new_Uk(data=labels)
    return centroid_values, labels  

# Function to calculate sum of square of errors
def calculate_SSE(labels, u_k):
    error_list = []
    for centriod_set, centriod in zip(labels, u_k):
        error = 0
        for elem_in_class in centriod_set:
            # print(f'data set elem {elem_in_class}, data of each centriod{centriod}')
            error += sum((np.array(elem_in_class)-centriod)**2)
        error_list.append(error)
    return sum(error_list)

# Function to plot elbow point given SSEs and K values
def plot_elbow_point(K_list,sse_list):
    plt.plot(K_list, sse_list) 
    plt.xlabel('Values of K') 
    plt.ylabel('SSE') 
    plt.title('The Elbow Method') 
    plt.show()
 

# Main program starts here
# Initialize variables

num_k = 10         # n_init: the number of initial guesses for the centroids
num_of_features=2  # Number of features
N=100              # Number of samples
sse_for_k=[]       # List to store SSEs for all runs for one value of K
sse_list=[]        # List to store min SSE value for each K
k_list=[]          # List of K values
num_of_runs=10     # Number of runs to be executed for each K

# Generate data for clustering
x_data, y_labels = generate_data(n_samples=N, n_features=num_of_features, random_state=0, centers=3)

# Run Kmeans 'num_of_runs' times for every K and find the lowest SSE for each K
for kval in range(1,num_k+1):
    
    # Run 'num_of_runs' times for each K and store min SSE value for that K
    for run in range(1,num_of_runs+1): 
        print("\nFor K ->",kval,"and Run ->",run)
        u_k, labels = k_means(x=x_data, K=kval,n_init=num_k) 
        plot(x=x_data, centroids=u_k)  
        
        #Store SSE values for all runs for a single value of k
        sse_for_k.append(calculate_SSE(labels, u_k))
    
    #Store the min value of SSE for a K
    sse_list.append(min(sse_for_k))
    k_list.append(kval)

# Plot Elbow point given list of errors and Ks
plot_elbow_point(k_list,sse_list)
```
