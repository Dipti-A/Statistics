##  (a) Consider the random number generation functions in NumPy,
Generate a set of random numbers using the function rand and plot its histogram. 

```python
vals = random.rand(100000)
data = pd.Series(vals)
data.plot.hist(data, bins=10)
```

### What is the shape of this histogram and why?
Using rand we see that the random vaules are generated, between 0 and 1 and from a uniform distribution, that is the frequency of each vaule is spread uniformally across all bins. 

### Investigate how the shape of the histogram is affected by the number of random numbers you have generated.
The larger the amount of vaules, that represent events, the higher the probablity that the frequency in all bins are exactly uniform, i,e, the shape is flat. So, 1000 vaules compared to 10,000 will both show uniform distribution where 10,000 is more likely to be uniform over all bins. Both experiments can be consired uniformlly distibuted. For very small number of events e.g. 10, the, we do n't have enough information to make an informed decsion about the shape or distribution. 

### Instead of using rand (which corresponds to a uniform distribution), generate numbers using some other distribution and plot a histogram. What is the shape now? For instance, with normal, the normal (or Gaussian) distribution, you should get the familiar bell shape

```python
vals = np.random.geometric(p=0.2, size=1000)
data = pd.Series(vals)
data.plot.hist(data, bins=100)
```

```python
vals = np.random.geometric(p=0.5, size=1000)
data = pd.Series(vals)
data.plot.hist(data, bins=100)
```

Geometric distribution.
This shows the number of events needed to achive a first 'True' or 'Success' event. The histrogram is skewed to the right, where the lower the p vaule the longer the tail, which means that more fauilres are more frequent. A p vaules of 1 results in every trail in an event being a success.

## (b) Modeling a student at an exam

Write a Python function that simulates that the student answers a single question either correctly or incorrectly. The function should return a Boolean value (that is, True or False) that says whether the question was answered correctly. You can assume that the probability of a correct answer is a given parameter p_success.

```python
def success(p_success):
    event = np.random.rand()
    if event < p_success:
        return True
    else:
        return False
```

```python
success(p_success=0.5)
```

Run this function a few times and check that it seems to work correctly.



```python

data = [success(0.5) for i in range(1000)]
#print(data)
from collections import Counter
print(Counter(data))
```


## How many correctly answered questions?

```python
def exam_score(p_correct, n_instances):
    """
    Runs for 10,000 iterations.
    Returns results for n_instances with P vaule for 10 thousand exams.
    """
    results = []
    for i in range(10000):
        data = [success(p_correct) for i in range(n_instances)]
        results.append(data)
    #s = np.random.binomial(n_instances, p_correct)
    return results
```

```python
data = exam_score(p_correct=0.8, n_instances=20)
```

## Counts the number of Successful questions per exam.

```python
from collections import Counter
counter_data = []
for elem in data:
    data_c = Counter(elem).get(1)
    counter_data.append(data_c)
```

## Plot the histogram for p vaules 0.8 for n_instances 10 and exam score 10,000 times.

```python
data = pd.Series(counter_data)
data.plot.hist(data, bins=10)
```

Since there is a high p value, .80 then we expect a high number of students that score well in the exam compared to ones that score lower.

# (c) The persistent student

```python
def number_of_attempts(p_pass, n):
    results = []
    for i in range(n):
        count = 0
        while not success(p_pass):
            count += 1
        results.append(count)
    return results
```

```python
attempts = number_of_attempts(0.4, 10000)
```

```python
attempts_data = pd.Series(attempts)
attempts_data.plot.hist(attempts_data, bins=10)
```

The p vaule is .40, 60 percent take about 0-2 attempts to get a pass. With a high P vaule we expect that more students pass on their first attempt and with a low p vaules students need more attempts to pass.
